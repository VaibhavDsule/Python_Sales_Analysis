```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv("Diwali Sales Data.csv ", encoding="unicode_escape")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
#drop unrelated/Blank column
```

```python
df.info()
```

```python
#Check for null values
df.isnull().sum()
```

```python
# Drop Null values
df.dropna(inplace=True)
```

```python
# Show row and columns
df.shape
```

```python
df.info()
```

```python
#Change the Data type Of columns
df['Amount']=df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
# Rename Column Header
df.rename(columns={'Cust_name' : 'Customer_name'})
```

```python
df.columns
```

```python
df.describe()
```

```python
#use describe function for specific columns
df[['Age','Orders','Amount']].describe()
```

# Gender

```python
df.columns
```

```python
sns.set_theme(style="darkgrid")
```

```python
sns.countplot(x='Gender',data=df)
plt.show()
```

```python

ax=sns.countplot(x='Gender',data=df)

for bar in ax.containers:
    ax.bar_label(bar)

```

```python
Sales_gn=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
Sales_gn
```

```python
Sales_gn=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=Sales_gn)
plt.show()
```

**From above graph we can see that most of the buyer are female and even the purchasing power of the female aregrater than men**

# Age

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Total amount vs Age Group
Sales_age= df.groupby(['Age Group','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_age,x='Age Group', y='Amount',hue='Gender')

```

**From above the chart we can say that most of the buyer age group between 26-35 Female**

# State

```python
df.columns
```

```python
#Total Order vs Top 10 Sate 
Order_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=Order_state,x='State',y='Orders')
```

```python
#Total amount vs Top 10 state 
Sales_sate=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=Sales_sate,x='State',y='Amount')
```

**From above graph we can see that most of the order and total sales/amount are from Uttarpradesh, Maharashtra and Karnataka respectively**

# Marital Satus

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Marital_Status',hue='Gender')
sns.set(rc={'figure.figsize':(5,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_ms=df.groupby(['Marital_Status',"Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_ms,x='Marital_Status',y='Amount',hue='Gender')
```

**From Above Chart we can se that most of the buyers are married(Womens) and they have high purchasing power**

# Occupation

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_occ,x='Occupation',y='Amount')
```

# Product category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data=df,x='Product_Category')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Total amount vs Product Category
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales_pc,x='Product_Category',y='Amount')
```

```python
#Total order vs Product id
sns.set({'figure.figsize':(20,5)})
Order_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=Order_pid,x='Product_ID',y='Orders')
```
